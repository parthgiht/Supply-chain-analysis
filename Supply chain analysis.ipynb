## Supply chain analysis

- The dataset provides a comprehensive view of **supply chain management** by incorporating various aspects of product lifecycle and logistics. It includes product details such as type, SKU (unique identifier), price, and availability, along with sales data like the number of products sold and revenue generated. Customer insights are captured through demographics, while inventory status is reflected in stock levels and lead times.

- The dataset also highlights order management, including quantities ordered, shipping times, carriers, and costs. It further extends to supplier details, such as names and locations, and includes production data like volumes, manufacturing lead times, and costs. Quality assurance is represented by inspection results and defect rates.

- Additionally, transportation information, including modes, routes, and associated costs, provides a detailed overview of logistics. This dataset is ideal for analyzing supply chain efficiency, identifying bottlenecks, and optimizing logistics and manufacturing processes.

## Columns Description

1. **Product type** : Type of the product (e.g., haircare, skincare).
2. **SKU** : Stock Keeping Unit identifier for the product.
3. **Price** : Price of the product.
4. **Availability** : Quantity available in stock.
5. **Number of products sold** : Total units sold.
6. **Revenue generated** : Total revenue earned from the product.
7. **Customer demographics** : Customer information like gender or classification.
8. **Stock levels** : Current stock level of the product.
9. **Lead times** : Time required to deliver the product.
10. **Order quantities** : Number of items ordered.
11. **Shipping times** : Time taken for shipping.
12. **Shipping carriers** : Name of the carrier handling the shipping.
13. **Shipping costs** : Cost of shipping the product.
14. **Supplier name** : Name of the supplier providing the product.
15. **Location** : Geographic location related to the supplier or product.
16. **Lead time** : Additional information on lead time (may differ from other lead times).
17. **Production volumes** : Total production output.
18. **Manufacturing lead time** : Time required for manufacturing.
19. **Manufacturing costs** : Cost of manufacturing the product.
20. **Inspection results** : Status or results of quality checks.
21. **Defect rates** : Percentage or rate of defective items.
22. **Transportation modes** : Mode of transport used (e.g., road, air).
23. **Routes** : Specific routes used for transportation.
24. **Costs** : Overall costs related to production, transportation, etc

## Average lead time and Average Manufacturing Costs for all products of the company

```python
Supply_df.head()
```

```python
# Groupby on product type with lead time
Avg_lead_time = Supply_df.groupby('Product type')['Lead time'].mean().reset_index()
```

```python
# Groupby on product type with manufacturing  costs
Avg_manufacturing_costs = Supply_df.groupby('Product type')['Manufacturing costs'].mean().reset_index()
```

```python
# Let's merge both Avg
Result = pd.merge(Avg_lead_time, Avg_manufacturing_costs, on = 'Product type')
```

```python
# Rename columns
Result.rename(columns = {'Lead time' : 'Average lead time', 'Manufacturing costs' : 'Average manufacturing costs'}, inplace = True)
```

```python
Result
```

## Analyzing SKU's (Stock keeping units)

**SKU's mean?**

- Imagine you have a large library with thousands of books. Each book is different, with its own title, author, and genre. To make it easier to keep track of all the books, the library assigns each book a unique code, like a secret label only the library knows. This code helps librarians quickly find out how many copies of a book are available, where they are located, or if they need to order more.This secret number is called SKU.
